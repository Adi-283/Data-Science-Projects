```python
covid_new['Total case'] = covid_new['Total Confirmed cases (Indian National)'] + covid_new['Total Confirmed cases ( Foreign National )']
```

```python
covid_new
```

```python
Total = covid_new['Cured'] + covid_new['Death']
```

```python
Total
```

```python
covid_new['Active case'] = covid_new['Total case'] - Total
```

```python
state_wise_case = covid_new.groupby('Name of State / UT')['Total case'].sum()
state_wise_case
```
